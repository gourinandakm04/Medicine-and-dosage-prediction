```python
#split dataset in features and target variable
feature_cols = ['Morning','Afternoon','Evening','Night','111','110','101','100','0-1-1','0-1-0','0-0-1','222','220','200','0-2-2','0-2-0','2-0-0','0-0-2','1/2-0-0','0-1/2-0','0-0-1/2','1-1-1/2','1-1/2-1','1/2-1-1','1/2-1/2-1','1-1/2-1/2','1/2-1/2-1/2','1/2-1/2-0','1/2-0-1/2','0-1/2-1/2','1-0-1/2','1-1/2-0','0-1/2-1','1/2-1-0','BD','TID','SOS-stat','stat','w/k','/2wk','Thrice','Twice','Weekly','half','0-1/2','/day']
X = df[feature_cols ] # Features
label=['count']
y = df[label]
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.000000001, random_state=0) # 70% training and 30% test
```

```python
 # Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)


#Predict the response for test dataset
#y_pred = clf.predict([[1, 0, 0, 1, 1, 0, 1, 1]])
```
